```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

```python
def funcStoreWeeklyScores(df):
    lst_rfp_t = []
    lst_rfp_l = []
    lst_rfp_c = []
    for index, row in df.iterrows():
        rfp_t = []
        rfp_l = []
        rfp_c = []
        for colname in list(df.columns):
            if colname.startswith('RFP Tech Ability wk'):
                rfp_t.append(row[colname])
            elif colname.startswith('RFP Learnability wk'):
                rfp_l.append(row[colname])
            elif colname.startswith('RFP Communicability wk'):
                rfp_c.append(row[colname])
            else:
                continue
        
        lst_rfp_t.append(rfp_t)
        lst_rfp_l.append(rfp_l)
        lst_rfp_c.append(rfp_c)
    return lst_rfp_t, lst_rfp_l, lst_rfp_c
            
```

```python
lst_rfp_t, lst_rfp_l, lst_rfp_c = funcStoreWeeklyScores(df)
```

```python
def funcEvaluateSlopeIntercept(lst, string, char):
    cols = []
    for index in range(len(lst[0])):
        cols.append(string + str(index+1))
    dataframe = pd.DataFrame(lst, columns=cols)
    dataframe[[char + "_" + 'slope', char + "_" +'intercept','r','p','se']] = dataframe.apply(lambda x: linregress(range(1,len(x)+1), x.values), axis=1).apply(pd.Series)
    new_df = dataframe[[char + "_" +'slope',char + "_" +'intercept']]
    return new_df
```

```python
# DataFrame Containing the values of slope and intercept for 'RFP Tech Ability'
df1 =  funcEvaluateSlopeIntercept(lst_rfp_t, "RFP Tech Ability wk", 'T')

# DataFrame Containing the values of slope and intercept for 'RFP Learnability'
df2 =  funcEvaluateSlopeIntercept(lst_rfp_l, "RFP Learnability wk", 'L')

# DataFrame Containing the values of slope and intercept for 'RFP Communicability'
df3 =  funcEvaluateSlopeIntercept(lst_rfp_c, "RFP Communicability wk",'C')
```

```python
concatenated_df = pd.concat([df, df1, df2, df3], axis=1)
```
