<h3>Importing Libraries</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,classification_report
```

<h3>Loading the Dataset</h3>

```python
tmr_data_df = pd.read_csv('/home/hp/Desktop/TMR-EDA/Data/encoded_tmr_data.csv')
tmr_data_df.head()
```

<h3>Renaming the columns</h3>

```python
tmr_data_df = tmr_data_df.rename(columns={ 
                                           'RFP Tech Ability \n(Out of 2)':'RFP Tech Ability',
                                           'RFP- Learnability \n(Out of 2)':'RFP- Learnability',
                                           'RFP-Communicability \n(Out of 1)':'RFP-Communicability',
                                           'Trial Techability Score (Out of 2.0)':'Trial Techability Score',
                                           'Trial Learnability Score (Out of 2.0)':'Trial Learnability Score',
                                           'Trial Communicability Score (Out of 1.0)':'Trial Communicability Score'
                                        })
tmr_data_df.head(3)
```

```python
# y = training_data['TechCategory']
# X = training_data.drop(columns='TechCategory', axis=1)
y = tmr_data_df['TechCategory']
```

```python
columns = ['RFP Last Internal Rating', 'RFP % Present', 'Trial Last Internal Rating', 'Trial % Present', 'RFP Avg TRACK Score', 'RFP Last Status', 'Trial Avg TRACK Score', 'RFP Tech Ability', 'RFP- Learnability', 'RFP Last TRACK Score', 'Trial Techability Score', 'Trial Learnability Score', 'RFP-Communicability', 'Last Graduation', 'Trial Communicability Score', 'Year']
print(columns)
```

<h3>Applying the model on different combinations of features</h3>

```python
def permute(y,columns,classifier,classifer_name):
    """
        Description:
            This function is used to apply the classification model on different set of features by 
            eliminating the last feature(least important feature) at the end of each iteration, 
            and storing the result in a .txt file.
        Parameters:
            columns: List of features in descending order of there importance.
            classifier: The object of classification model to be applied.
            classifer_name: The name of the classification algorithm used.
        Return:
            None                   
    """
    with open(r'/home/hp/Desktop/TMR-EDA/Data/classification_report.txt', 'a') as f:
        f.write('Model : ' + classifer_name + '\n')
        for i in range(len(columns)):
            data_column_list=columns
            training_data = tmr_data_df.filter(data_column_list, axis=1)
            X_train, X_test, y_train, y_test = train_test_split(training_data, y,test_size=0.15, random_state=0)
            classifier.fit(X_train,y_train)
            y_pred = classifier.predict(X_test)
            report = classification_report(y_test, y_pred, output_dict=True)
            df = pd.DataFrame(report).transpose()       
            dfAsString = df.to_string(header=True, index=False)
            f.write(','.join(columns) + '\n')
            f.write(dfAsString + '\n' + '\n')
            columns = columns[:-1]
        f.write('*'*200 + '\n')
```

<h3>Model Building</h3>

<h2>Logistic Regression</h2>

```python
logistic_regression = LogisticRegression(random_state=0)
permute(y,columns,logistic_regression,'LogisticRegression')
```

<h2>Random Forest</h2>

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators=20, random_state=0)
permute(y,columns,RF,'RandomForestClassifier')
```

<h2>Support Vector Classifier</h2>

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='rbf')
permute(y,columns,svclassifier,'SupportVectorClassifier')
```
