**Dropping the rows from df1 for with the Admit Id is empty**

```python
df1 = df1.dropna(axis=0, subset=['Admit Id'])
```

**Finding the Duplicate Rows**

```python
# Selecting duplicate rows except first
# occurrence based on all columns
duplicate = df1[df1.duplicated()]
print("Duplicate Rows : ")
duplicate
```

```python
df1.shape
```

**Dropping the rows from df1 for with the Admit Id is '-'**

```python
# get names of indexes for which
# column Admit Id has value '-'
index_names = df1[ df1['Admit Id'] == '-' ].index
  
# drop these row indexes
# from dataFrame
df1.drop(index_names, inplace = True)
```

```python
df1.shape
```

**Keeping only those rows wherein the Admit Id starts with CIC**

```python
df1 = df1[df1["Admit Id"].str.startswith("CIC")]
```

```python
df1.shape
```

**Removing rows wherein Regularity Status is 'Remap'**

```python
index_names = df1[ df1['Regularity Status'] == 'Remap' ].index
df1.drop(index_names, inplace = True)
```

```python
df1.shape
```
